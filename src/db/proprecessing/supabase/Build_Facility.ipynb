```python
import os
from supabase import create_client, Client

url: str = "********"
key: str = "********"
supabase: Client = create_client(url, key)
```

```python
import pandas as pd

df1 = pd.read_csv('../disability_facilities/공동생활가정_경기도_위경도.csv')
df2 = pd.read_csv('../disability_facilities/단기거주시설_경기도_위경도.csv')
df3 = pd.read_csv('../disability_facilities/중증장애인거주시설_경기도_위경도.csv')
df4 = pd.read_csv('../disability_facilities/지체장애인시설_경기도_위경도.csv')
```

```python
df = pd.DataFrame()
df1['type'] = '공동생활가정'
df2['type'] = '단기거주시설'
df3['type'] = '중증장애인거주시설'
df4['type'] = '지체장애인시설'

df = pd.concat([df, df1])
df = pd.concat([df, df2])
df = pd.concat([df, df3])
df = pd.concat([df, df4])

df = df.reset_index(drop=True)

for i in df.index:
    df.loc[i, 'location'] = f"POINT({df.loc[i,'lon']} {df.loc[i,'lat']})"
    
df2 = df.drop(columns = ['일련번호', '시군구', '입소정원', '입소현원', '종사정원', '종사현원', 'lat', 'lon'])
df2
```

```python
df2.columns = ['name', 'address', 'phone', 'installed_at', 'corporate', 'type', 'location']
df2.phone = df2.phone.str.replace(' ', ',')
df2.address = df2.address.str.rstrip()
df2
```

```python
df2.installed_at = pd.to_datetime(df2['installed_at'])
df2["installed_at"] = df2["installed_at"].apply(lambda x: x.isoformat())
```

```python
dictionary = df2.to_dict(orient='records')
```

```python
dictionary
```

```python
data, count = supabase.table('disability_facilities').insert(dictionary).execute()
```
