```python
for i in df.index:
    df.loc[i, "location"] = f"POINT({df.loc[i, 'lon']} {df.loc[i, 'lat']})"
df2 = df.drop(columns=['lat', 'lon'])
df2['is_male'] = df2['male'] > 0
df2['is_female'] = df2['female'] > 0
df2['비상벨설치여부'].replace('Y', True, inplace=True)
df2['비상벨설치여부'].replace('N', False, inplace=True)
```

```python
df2.columns = ['name', 'address', 'is_emergency_bell', 'opening_hours', 'male_count', 'female_count', 'location', 'is_male', 'is_female']
```

```python
dictionary = df2.to_dict(orient='records')
```

```python
data, count = supabase.table('toilets').insert(dictionary).execute()
```
