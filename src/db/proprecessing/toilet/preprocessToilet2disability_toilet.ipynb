```python
import pandas as pd

df = pd.read_excel('경기도공중화장실정보.xlsx')
df
```

```python
columns_with_disability = [col for col in df.columns if '장애인' in col and '대변기' in col]
col = columns_with_disability + ['화장실명', '소재지도로명주소', '소재지지번주소', '비상벨설치여부', '개방시간', '개방시간상세']
df2 = df[col]
df2
```

```python
df2['male'] = df2['남성용-장애인용대변기수']
df2['female'] = df2['여성용-장애인용대변기수']

df2.loc[df2['개방시간상세'].isnull(), '개방시간상세'] = "00:00~00:00"

mask1 = df2['소재지도로명주소'].isnull()
df2.loc[mask1, '소재지도로명주소'] = df2.loc[mask1, '소재지지번주소']

mask2 = df2['male'] > 0
mask3 = df2['female'] > 0

df3 = df2.loc[mask2 | mask3].drop(columns = ['남성용-장애인용대변기수', '여성용-장애인용대변기수', '개방시간', '소재지지번주소'])
df3
```

```python
df3.to_csv('경기도 장애인 공중 화장실 정보.csv', index=False)
```
