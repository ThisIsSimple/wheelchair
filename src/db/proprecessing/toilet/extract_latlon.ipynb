```python
maps = googlemaps.Client(key=API_KEY)
df = pd.read_csv('경기도 장애인 공중 화장실 정보.csv')
df
```

```python
from tqdm import tqdm
lat, lon = [], []

for address in tqdm(df['소재지도로명주소']):
    try:
        geo_location = maps.geocode(address)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lon.append(geo_location['location']['lng'])    
    except:
        lat.append('')
        lon.append('')
        print(f"{address} 에러")
```

```python
df['lat'] = lat
df['lon'] = lon
df
```

```python
df.to_csv('경기도 장애인화장실 위치 정보.csv', index=False)
```
